```python
#!/usr/bin/python3.7

"""
Code to create county-level .csv files from restaurant
inspection data.

"""
# Last updated 7/16/2019 by doug.ray@starbanner.com

# built-in libraries
import csv
import datetime
import os.path
import re
import sys

# installed with pip
import pandas as pd
import numpy as np

# Create dataframe from files
colnames = [
    "county", "licnum", "sitename", "streetaddy", "cityaddy", "zip",
    "inspnum", "insptype", "inspdispos", "inspdate", "totalvio", 
    "highvio", "licid", "visitid"
    ]

colnums = [2, 4, 5, 6, 7, 8, 9, 12, 13, 14, 17, 18, 80, 81]

filepath18_1 = 'fy2018/1fdinspi_1819.csv'
filepath18_2 = 'fy2018/2fdinspi_1819.csv'
filepath18_3 = 'fy2018/3fdinspi_1819.csv'
filepath18_4 = 'fy2018/4fdinspi_1819.csv'
filepath18_5 = 'fy2018/5fdinspi_1819.csv'
filepath18_6 = 'fy2018/6fdinspi_1819.csv'
filepath18_7 = 'fy2018/7fdinspi_1819.csv'

df18_1 = pd.read_csv(
    filepath18_1,
    names=colnames,
    usecols=colnums,
    encoding="ISO-8859-1"
    )

df18_2 = pd.read_csv(
    filepath18_2,
    names=colnames,
    usecols=colnums,
    encoding="ISO-8859-1"
    )

df18_3 = pd.read_csv(
    filepath18_3,
    names=colnames,
    usecols=colnums,
    encoding="ISO-8859-1"
    )

df18_4 = pd.read_csv(
    filepath18_4,
    names=colnames,
    usecols=colnums,
    encoding="ISO-8859-1"
    )

df18_5 = pd.read_csv(
    filepath18_5,
    names=colnames,
    usecols=colnums,
    encoding="ISO-8859-1"
    )

df18_6 = pd.read_csv(
    filepath18_6,
    names=colnames,
    usecols=colnums,
    encoding="ISO-8859-1"
    )

df18_7 = pd.read_csv(
    filepath18_7,
    names=colnames,
    usecols=colnums,
    encoding="ISO-8859-1"
    )

df18_all = pd.concat(
    [df18_1, df18_2, df18_3, df18_4, df18_5, df18_6, df18_7], ignore_index=True
    )
```

```python
df18_novios = df18_all.loc[df18_all['totalvio'] == 0]
df18_nohi = df18_all.loc[df18_all['highvio'] == 0]
```

```python
gh_markets = ['Alachua', 'Okaloosa', 'Walton', 'Volusia', 'Flagler',
             'Duval', 'Palm Beach', 'Sarasota', 'Marion', 'Lake',
             'Manatee', 'Polk', 'St. Johns']

# Make .csv files for each county in a Gatehouse market; this one
# showing restaurants that had inspections with no violations found.

df_okaloosa = df18_novios.loc[df18_novios['county'] == 'Okaloosa']
df_okaloosa.to_csv('county_level/okaloosa_no_vios.csv')

df_alachua = df18_novios.loc[df18_novios['county'] == 'Alachua']
df_alachua.to_csv('county_level/alachua_no_vios.csv')

df_walton = df18_novios.loc[df18_novios['county'] == 'Walton']
df_walton.to_csv('county_level/walton_no_vios.csv')

df_volusia = df18_novios.loc[df18_novios['county'] == 'Volusia']
df_volusia.to_csv('county_level/volusia_no_vios.csv')

df_flagler = df18_novios.loc[df18_novios['county'] == 'Flagler']
df_flagler.to_csv('county_level/flagler_no_vios.csv')

df_duval = df18_novios.loc[df18_novios['county'] == 'Duval']
df_duval.to_csv('county_level/duval_no_vios.csv')

df_palm = df18_novios.loc[df18_novios['county'] == 'Palm Beach']
df_palm.to_csv('county_level/palm_no_vios.csv')

df_sarasota = df18_novios.loc[df18_novios['county'] == 'Sarasota']
df_sarasota.to_csv('county_level/sarasota_no_vios.csv')

df_marion = df18_novios.loc[df18_novios['county'] == 'Marion']
df_marion.to_csv('county_level/marion_no_vios.csv')

df_lake = df18_novios.loc[df18_novios['county'] == 'Lake']
df_lake.to_csv('county_level/lake_no_vios.csv')

df_manatee = df18_novios.loc[df18_novios['county'] == 'Manatee']
df_manatee.to_csv('county_level/manatee_no_vios.csv')

df_polk = df18_novios.loc[df18_novios['county'] == 'Polk']
df_polk.to_csv('county_level/polk_no_vios.csv')

df_stjohn = df18_novios.loc[df18_novios['county'] == 'St. Johns']
df_stjohn.to_csv('county_level/stjohn_no_vios.csv')
```
